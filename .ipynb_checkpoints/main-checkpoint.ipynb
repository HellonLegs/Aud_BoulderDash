```c++
#include "Map.hpp"
#include <algoviz/SVG.hpp>
#include <iostream>
#include <string> 
#include <chrono>
using namespace std;
SVG view = SVG(480, 256);;
```

```c++
bool haupt(int lvl){
    std::chrono::time_point<std::chrono::system_clock> previousTimePoint, timePointNow;
    view.clear();
    Map feld = Map(lvl, &view);
    string text;
    int points = 0;
    int dx;
    int dy;
    string last = "";
    int fertig = 0;
    previousTimePoint = std::chrono::high_resolution_clock::now();
    
    while(fertig == 0 && last != "q"){
        dx = 0;
        dy = 0;
        last = view.lastKey();
        if(last == "ArrowUp"){
            dx = 0;
            dy = -1;
        }else if(last == "ArrowRight"){
            dx = 1;
            dy = 0;
        }else if(last == "ArrowDown"){
            dx = 0;
            dy = 1;
        }else if(last == "ArrowLeft"){
            dx = -1;
            dy = 0;
        }
        int p = feld.movePlayer(dx, dy, last);
        if(p > 0){
            points++;
        }
        text = "Du hast ";
        text += to_string(points);
        text += " von 6 benoetigten Diamanten gesammelt";
        view.setStatusMsg(text);
        feld.moveEnemys();
        feld.checkFalling();
        timePointNow = std::chrono::high_resolution_clock::now();
        std::chrono::duration<double> duration = (timePointNow - previousTimePoint);
        if(feld.checkDeath()){
            fertig = -1;
        }else if(duration.count() > 150000.0){
            fertig = -2;
        }
        else if(p < 0 && points >= 6){
            fertig = 1;
        }
        feld.drawAll();
        AlgoViz::sleep(50);
    }
    feld.checkFalling();
    feld.drawAll();
    switch (fertig){
        case -2 :   text = "Die Zeit ist abgelaufen.";
                    break;
        case -1 :   text = "Du bist gestorben.";
                    break;
        case 1 :    text = "Du hast mit ";
                    text += to_string(points);
                    text += " Diamanten gewonnen!";
                    break;
        case 0 :    text = "ABBRUCH!";
                    break;
    }
    view.setStatusMsg(text);
    AlgoViz::sleep(5000);
    if(fertig >= 0){
        return true;
    }
    return false;
}
```

```c++
int lives = 3;
while(!haupt(0) % lives > 0){
    lives--;
}
lives = 3;
while(!haupt(1) % lives > 0){
    lives--;
}
```
